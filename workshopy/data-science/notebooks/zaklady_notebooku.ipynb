# Jupyter Notebook — základy

Jupyter Notebook je interaktivní prostředí, které kombinuje **kód**, **text** a **vizualizace** v jednom dokumentu.

Notebook se skládá z **buněk** (cells). Existují dva hlavní typy:

| Typ | Popis |
|-----|-------|
| **Code** | Spustitelný kód (Python) |
| **Markdown** | Formátovaný text (nadpisy, seznamy, tabulky, obrázky, vzorce) |

Tato buňka je **Markdown**. Následující buňka je **Code**.

````python
a = 3 + 3
a
````

## Funkce

````python
def pozdrav(jmeno):
    return f"Ahoj, {jmeno}!"

pozdrav("svete")
````

````python
def fibonacci(n):
    """Vrátí prvních n členů Fibonacciho posloupnosti."""
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib[:n]

fibonacci(10)
````

## Importování knihoven

Python má bohatý ekosystém knihoven. Importujeme je pomocí `import`.

````python
import math

print(f"π = {math.pi}")
print(f"e = {math.e}")
print(f"√2 = {math.sqrt(2)}")
print(f"5! = {math.factorial(5)}")
````

````python
import numpy as np

# NumPy pole
pole = np.array([1, 2, 3, 4, 5])
print(f"Pole:    {pole}")
print(f"Průměr:  {pole.mean()}")
print(f"Suma:    {pole.sum()}")
print(f"Pole*2:  {pole * 2}")
````

## Vizualizace

````python
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Goniometrické funkce')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Sloupcový graf
jazyky = ['Python', 'C', 'JavaScript', 'Java', 'Go']
popularita = [85, 60, 75, 55, 40]

plt.bar(jazyky, popularita, color=['#3776ab', '#555555', '#f7df1e', '#f89820', '#00add8'])
plt.ylabel('Popularita')
plt.title('Programovací jazyky')
plt.show()
````

## Užitečné klávesové zkratky

Notebook má dva režimy:

- **Edit mode** (zelený okraj) — stiskněte `Enter`

### Command mode

| Zkratka | Akce |

| Zkratka | Akce |
