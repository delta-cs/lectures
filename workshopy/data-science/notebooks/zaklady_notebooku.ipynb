# Jupyter Notebook — základy

Jupyter Notebook je interaktivní prostředí, které kombinuje **kód**, **text** a **vizualizace** v jednom dokumentu.

Notebook se skládá z **buněk** (cells). Existují dva hlavní typy:

| Typ | Popis |
|-----|-------|
| **Code** | Spustitelný kód (Python) |
| **Markdown** | Formátovaný text (nadpisy, seznamy, tabulky, obrázky, vzorce) |

Tato buňka je **Markdown**. Následující buňka je **Code**.

## Proměnné

````python
a = 3 + 3
a
````

Proměnné **přežívají mezi buňkami** — jakmile proměnnou definujete, můžete ji používat v jakékoliv další buňce.

````python
b = a * 2
b
````

````python
# To display multiple values, use print()
print("a =", a)
print("b =", b)
print("a + b =", a + b)
````

## Funkce

````python
def greet(name):
    return f"Hello, {name}!"

greet("world")
````

````python
def fibonacci(n):
    """Return the first n Fibonacci numbers."""
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib[:n]

fibonacci(10)
````

Python má bohatý ekosystém knihoven. Importujeme je pomocí `import`.

## Vizualizace

````python
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric functions')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Bar chart
languages = ['Python', 'C', 'JavaScript', 'Java', 'Go']
popularity = [85, 60, 75, 55, 40]

plt.bar(languages, popularity, color=['#3776ab', '#555555', '#f7df1e', '#f89820', '#00add8'])
plt.ylabel('Popularity')
plt.title('Programming languages')
plt.show()
````
